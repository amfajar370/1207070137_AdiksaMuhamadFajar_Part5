```python
import matplotlib.pyplot as plt

from skimage import data
from skimage.io import imread
from skimage.color import rgb2gray
from skimage.util import invert
import numpy as np
```

```python
#percobaan 1
astronautImage = data.astronaut()
cameraImage = data.camera()

astroCropped = astronautImage.copy()
astroCropped = astroCropped[0:256,64:320]

cameraCropped = cameraImage.copy()
cameraCropped = cameraCropped[64:320,128:320]

fig, axes = plt.subplots(2, 2, figsize = (12, 12))
ax = axes.ravel()

ax[0].imshow(astronautImage)
ax[0].set_title("Citra Input 1")
ax[1].imshow(cameraImage, cmap='gray')
ax[1].set_title("Citra Input 2")
ax[2].imshow(astroCropped)
ax[2].set_title("Citra Output 1")
ax[3].imshow(cameraCropped, cmap='gray')
ax[3].set_title("Citra Output 2")

fig.tight_layout
plt.show()
```

```python
import matplotlib.pyplot as plt
import cv2

from skimage import data
from skimage.io import imread
from skimage.color import rgb2gray
from skimage.util import invert
import numpy as np
```

```python
loadImage = imread(fname="pamungkas.jpg")
loadImagecropped = loadImage.copy()
loadImagecropped = loadImagecropped[140:500,240:720]

fig, axes = plt.subplots(1, 2, figsize = (10, 10))
ax = axes.ravel()

ax[0].imshow(loadImage)
ax[0].set_title("Citra Input 1")
ax[1].imshow(loadImagecropped)
ax[1].set_title("Citra Output 2")

fig.tight_layout
plt.show()



```

```python
inv = invert(loadImagecropped)
print('Shape Input : ', loadImagecropped.shape)
print('Shape Output : ',inv.shape)

fig, axes = plt.subplots(2, 2, figsize=(12, 12))
ax = axes.ravel()

ax[0].imshow(loadImagecropped)
ax[0].set_title("Citra Input")

ax[1].hist(loadImagecropped.ravel(), bins=256)
ax[1].set_title('Histogram Input')

ax[2].imshow(inv)
ax[2].set_title('Citra Output (Inverted Image)')

ax[3].hist(inv.ravel(), bins=256)
ax[3].set_title('Histogram Output')
```

```python
imggray = rgb2gray(loadImagecropped) # mengubah citra menjadi grayscale

print('img2gray shape = ',imggray.shape) # menampilkan dimensi citra

fig, axes = plt.subplots(1, 2, figsize=(12, 12)) # membuat subplot dengan 2 baris dan 2 kolom dengan jarak 12
ax = axes.ravel()

ax[0].imshow(loadImagecropped) # menampilkan hasil citra cropped 
ax[0].set_title("img2 RGB Image")
ax[1].imshow(imggray,cmap=plt.cm.gray) # menampilkan hasil citra cropped yg sudah grayscale
ax[1].set_title("img2 Grayscale Image")

fig.tight_layout() # mengatur tata letak subplot
plt.show() # menampilkan subplot
```

```python
copyCamera = imggray.copy().astype(float) # menyalin citra grayscale dan mengubah type datanya

m1,n1 = imggray.shape # untuk mendapatkan dimensi citra yang disalin
output1 = np.empty([m1, n1]) # membuat array kosong

for baris in range(0, m1-1): # melakukan pengolahan pada setiap piksel citra
    for kolom in range(0, n1-1):
        a1 = baris
        b1 = kolom
        output1[a1, b1] = copyCamera[baris, kolom] + 2 #memberi brightness pada output sebesar 2
        
fig, axes = plt.subplots(2, 2, figsize=(12, 12)) # membuat subplot dengan 2 baris dan 2 kolom dengan jarak 12
ax = axes.ravel()

ax[0].imshow(copyCamera, cmap='gray') # menampilkan citra dalam format keabuan
ax[0].set_title("Citra Input")

ax[1].hist(copyCamera.ravel(), bins=256) # menampilkan histogram 
ax[1].set_title('Histogram Input')

ax[2].imshow(output1, cmap='gray') # menampilkan citra hasil penambahan brightness
ax[2].set_title('Citra Output (Brightnes) + 2')

ax[3].hist(output1.ravel(), bins=256) # menampilkan histogramnnya
ax[3].set_title('Histogram Input')
```
